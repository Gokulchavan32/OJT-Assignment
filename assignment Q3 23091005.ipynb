```python
import numpy as np
import sympy as sp
#Q3
# Singular-value decomposition
np.random.seed(5)
A = np.random.randint(0, 10, size=(5, 5))
sp.Matrix(A)
```

```python
U,s,Vt = np.linalg.svd(A)
```

```python
sp.Matrix(np.round(U))
```

```python
s
```

```python
Sigma = np.zeros((A.shape[0], A.shape[1]))
sp.Matrix(Sigma)
```

```python
Sigma[:A.shape[1], :A.shape[0]] = np.diag(s)
sp.Matrix(Sigma)
```

```python
sp.Matrix(Vt)
```

```python
#Verification of SVD
B=U@Sigma@Vt
```

```python
sp.Matrix(B)
```

```python
C=np.round(B)
sp.Matrix(C)
```

```python
#rank 2 approximation
k = 1
Uk = U[:,:k] 
Sigmak = Sigma[:k,:k] 
Vtk = Vt[:k,:] 
Ak = Uk@Sigmak@Vtk
sp.Matrix(Ak)
```

```python
#rank 3 approximation
k = 3
Uk = U[:,:k] 
Sigmak = Sigma[:k,:k] 
Vtk = Vt[:k,:] 
Ak = Uk@Sigmak@Vtk
sp.Matrix(Ak)
```
